```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")

ratings.head()
```

```python
# Merge the movies with their ratings given by each user
movie_ratings = movies.merge(ratings, on='movieId').drop(['genres', 'timestamp'], axis=1)
movie_ratings.head()
```

```python
#Pivot the table on userId and movie title
movie_ratings = movie_ratings.pivot_table(index=['userId'], columns=['title'], values='rating')
movie_ratings.head()
```

```python
# Drop the movies where less than 10 users have rated the movie
movie_ratings_raw = movie_ratings.dropna(thresh=10, axis=1)
movie_ratings = movie_ratings_raw.fillna(0)
movie_ratings.head()
```

```python
#calculate karls pearsons correlation coefficient for each movie (closer the value the 1 the more similar the movies)
movie_similarity = movie_ratings.corr(method="pearson")
movie_similarity.head()
```

```python
# Return movies based on the movie and the user given by the user
def similar_movie(movie_name, user_rating):
    similar_score = movie_similarity[movie_name] * (user_rating - 2.5)
    return similar_score.sort_values(ascending=False)
```

```python
#Recommends 10 movies based on the user
def recommend_movies(user_id):
    
    #get the movies viewed by the user
    user = movie_ratings_raw.iloc[user_id].dropna().to_frame()

    similar_movies = pd.DataFrame()

    for row in user.iterrows():
        movie_name = row[0]
        rating = row[1].iloc[0]
        similar_movies = similar_movies.append(similar_movie(movie_name, rating), ignore_index=True)

    recommendations = similar_movies.sum().sort_values(ascending=False)

    #Drop the movies the user has already watched
    for row in user.iterrows():
        movie_name = row[0]
        recommendations = recommendations.drop(labels=movie_name)
    
    movieList = []
    i = 0
    for row in recommendations.iteritems():
        if i < 10:
            i += 1
            movieList.append(row[0])
        
    return movieList
```

```python
user_id = 300

df = pd.DataFrame()
recommended_movies = recommend_movies(user_id)

for movie in recommended_movies:
    df = df.append(movies.loc[movies['title'] == movie])

print("Recommended movies: ")
df.head(10)
```
